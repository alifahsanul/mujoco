```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
def get_latest_file(parent_path, ext):
    files = os.listdir(parent_path)
    paths = [os.path.join(parent_path, basename) for basename in files if basename.endswith(ext)]
    return max(paths, key=os.path.getctime)
```

```python
parent_folder = 'training_result'
save_freq = 20000

# get the latest folder
folders = os.listdir(parent_folder)
folders.sort()
folder = folders[-1]
# get the latest file
file = get_latest_file(os.path.join(parent_folder, folder), 'npy')

print(folder, file)

reward = np.load(file)
episode= np.arange(0, len(reward) * save_freq, save_freq)
# plot the reward
plt.plot(episode, reward)
plt.xlabel('Episodes')
plt.ylabel('Reward')
plt.title('Reward over episodes')
plt.show()
```
